Escenario de negocio

Una clínica privada quiere implementar un sistema automático que, a partir de datos médicos básicos de una paciente, estime el riesgo de diabetes.

El sistema será integrado en su software interno mediante una API REST que devuelva una predicción (0 o 1) basada en los datos ingresados.

El equipo de datos debe construir:

Un pipeline reproducible de Machine Learning

Persistencia del modelo entrenado

Un servicio desacoplado que lo consuma

Una API REST con validación robusta

Objetivo del proyecto

Construir un sistema de clasificación completo que:

Entrene un modelo de ML usando scikit-learn

Evalúe su desempeño correctamente

Persista el modelo entrenado

Exponga un endpoint /predict mediante FastAPI
